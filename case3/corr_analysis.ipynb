```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

df = pd.read_csv("Training Data_Case 3.csv")
df = df.rename(columns={'Unnamed: 0': 'date'})
df.set_index('date', inplace=True)


print(df.head())
```

```python
# Plot the returns for each asset
returns = (df / df.iloc[0]) - 1
ax = returns.plot(figsize=(10, 6))
ax.set_ylabel('Return (%)')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0))
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
ax.set_title('Correlation Matrix')
plt.show()
```

```python
# Compute the covariance matrix for returns data
cov_matrix_norm = returns.cov()
fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(cov_matrix_norm, annot=True, cmap='coolwarm', ax=ax)
ax.set_title('Variance Covariance Matrix')
plt.show()
```
